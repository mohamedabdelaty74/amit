##convert categorical variables

```python
ecommerce_data["OrderDate"]=pd.to_datetime(ecommerce_data["OrderDate"])
```

```python
ecommerce_data["Delivery Date"]=pd.to_datetime(ecommerce_data["Delivery Date"])
```

```python
ecommerce_data['Customer Gender']=ecommerce_data['Customer Gender'].astype('category')
```

#important

```python
##
assert ecommerce_data['Customer Gender'].dtype=='category'
##
```

```python
ecommerce_data['Location']=ecommerce_data['Location'].astype('category')
```

```python
ecommerce_data['Zone']=ecommerce_data['Zone'].astype('category')
```

```python
ecommerce_data['Delivery Type']=ecommerce_data['Delivery Type'].astype('category')
```

```python
ecommerce_data['Product Category'] = ecommerce_data['Product Category'].astype('category')
ecommerce_data['SubCategory'] = ecommerce_data['SubCategory'].astype('category')
ecommerce_data['Product'] = ecommerce_data['Product'].astype('category')
ecommerce_data['Status'] = ecommerce_data['Status'].astype('category')
```

```python
plt.figure(figsize=(10,6))
sns.histplot(ecommerce_data['Customer Age'],bins=20,kde=True,color='skyblue')
plt.title('distribution of customer age')
plt.xlabel('age')
plt.ylabel('frequency')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Customer Gender',data=ecommerce_data,palette='pastel')
plt.title('count of orders by genders')
plt.xlabel('gender')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Zone',data=ecommerce_data,palette='pastel')
plt.title('count of orders by Zone')
plt.xlabel('zone')
plt.ylabel('orders')
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.countplot(x='Location',data=ecommerce_data,palette='Set2')
plt.title('count of orders by location')
plt.xlabel('location')
plt.ylabel('orders')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(ecommerce_data.corr(),annot=True,cmap='coolwarm',linewidths=0.5)
plt.title('correlation matrix')
```

```python
ecommerce_data['shipping cost percent']=(ecommerce_data['Shipping Fee'] / ecommerce_data['Sale Price'])*100
```
